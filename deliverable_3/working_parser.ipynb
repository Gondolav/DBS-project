```python
with open("madrid_host_out.csv", 'r') as h_in, open("madrid_host_out_2.csv", 'w') as h_out:
    hosts = csv.reader(h_in, delimiter=';', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    hosts_out = csv.writer(h_out, delimiter=';', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    
    header = h_in.readline()
    h_out.write(header)

    # Transform the rest of the lines
    for host in hosts:
        vs = host[10][2:-2].split('\', \'')
        
        for v in vs:
            with open("verifications.csv", 'r') as vs_r:
                verifications = csv.reader(vs_r, delimiter=';', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                vid = -1
                size = 0
                
                for verif in verifications:
                    size += 1
                    if v == verif[1]:
                        vid = int(verif[0])
                        with open("has_verif.csv", 'a') as ns:
                            ns.write("{};{}\n".format(vid, host[0]))
            if vid < 0:
                vid = size + 1
                with open("verifications.csv", 'a') as ns:
                    ns.write("{};{}\n".format(vid, v))
                with open("has_verif.csv", 'a') as ns:
                    ns.write("{};{}\n".format(vid, host[0]))
        hosts_out.writerow(host[:-1])
```

```python
def parse_price(price):
    if price == 'N/A' or price == '':
        return 'NULL'
    return float(price.replace("$", "").replace(",", ""))
```
